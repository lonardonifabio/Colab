```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

#units = layers of neural network
#input_shape = data input for layer
l0 = Dense(units=1, input_shape=[1])
model = Sequential([l0])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([1,2,3,4,5,6], dtype=float)
ys = np.array([0,3,5,7,9,11], dtype=float)

model.fit(xs, ys, epochs=500) #Y=2X-1
```

```python
print(model.predict([6]))
```

```python
print("Here is what I learned: {}".format(l0.get_weights())) #Y=2X-1
```
